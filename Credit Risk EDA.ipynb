```python
# Load the dataset
import pandas as pd
df = pd.read_csv("credit_risk_dataset.csv")

# Dsiplay top 5 rows in the data 
print(df.head())
```

```python
print(df.duplicated().sum())
```

```python
# Checking for duplicate rows if they are actually duplicates in the dataset
duplicate_rows = df[df.duplicated()]
duplicate_rows
```

```python
#Removing employemnt length outlier:emp_length cannot be greater than 60 and person_age cannot be greater than emp_length

df= df[df['person_emp_length'] <= 60]
df = df[df['person_emp_length'] < df['person_age']]
df.shape
```

```python
# Handling age outliers in the dataset by capping: by replacing the outlier ages with the median age
median_age = df[df['person_age'] <= 100]['person_age'].median()

# Replace any age over 100 with calculated median age
df.loc[df['person_age'] > 100, 'person_age'] = median_age
```

```python
# Handling null values
# In data analytics, missing data handling is called imputation
# Replacing the null values in the loan_interest rate column with median int_rate

median_int_rate = df['loan_int_rate'].median()
df['loan_int_rate'] = df['loan_int_rate'].fillna(median_int_rate)
```

```python
# Check if there are any null values in emp_length column
null_count_emp_length = df['person_emp_length'].isnull().sum()
print(null_count_emp_length)
```

```python
# Verify the dataset is clean
print(" POST CLEANING CHECK  ")
print("Missing values left: ", df.isnull().sum())
print(df.shape)
```
